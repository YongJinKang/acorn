EDA과정에서 PCA(새로운,의미있는 컬럼을 추가) 클래스구간화 등  정규화 결측치이상치 제거
데이터 탐색 단계에서 데이터 셋 구경, 결측처리, 널 값 처리, 처리된 데이터 이용해서 가공,(가공전 가공 후 데이터) 시각화도 필요,
독립변수와의 상관분석을 하여서 의미있는 변수들 뽑아내기



```python
#성별 인코딩
# LabelEnconder
from sklearn.preprocessing import LabelEncoder
features = ["Sex"]#, "Cabin"]
encoder = LabelEncoder() 
#===================fit() predict() 학습: fit에서 학습완료. 모든 작업 끝
#-------------------fit() transform() fit_transform() :: feature 변환(transform) :여기서 fit은 학습완료가 아니다. 모양만 맞춰놨고
# 변화는 transform할 때 실제로 일어난다.
# encoder.fit(df_train["Sex"])
# encoder.fit(df_train[["Sex", 'cabin']])
#---------------------------------인코딩한다 = Object -> 수치화(남: 0 , 여: 1)
for feature in features:
    encoder.fit(df_train[feature])
    df_train[feature] = encoder.transform(df_train[feature])
print(df_train["Sex"])
```

### 학습

```python
from sklearn.model_selection import train_test_split
```

```python
y_data = df_train.Survived
X_data = df_train.drop('Survived', axis=1)
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size = 0.3, random_state= 11, shuffle=True)


from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score 
dt_model = DecisionTreeClassifier()
# 모델 준비 완료
```

```python
dt_model.fit(X_train, y_train)
model_pred = dt_model.predict(X_test)
score = accuracy_score(y_test, model_pred )
print(score)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score
```

```python
print("정확도 ",accuracy_score(y_test, model_pred))
print("F1:",f1_score(y_test, model_pred))
print("정밀도: ",precision_score(y_test, model_pred))
print("재현률(민감도):",recall_score(y_test, model_pred))
print("ROC:",confusion_matrix(y_test, model_pred))
print("ROC_AUC:",roc_auc_score(y_test, model_pred))
```
