```python
from sklearn.datasets import load_breast_cancer
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import GridSearchCV, KFold, StratifiedKFold


def custom_cls_eval_(y, pred):
    print("정확도 ",accuracy_score(y, pred))
    print("F1:",f1_score(y, pred))
    print("정밀도: ",precision_score(y,pred))
    print("재현률(민감도):",recall_score(y, pred))
#     print("ROC:",confusion_matrix(y, pred))
    print("ROC_AUC:",roc_auc_score(y,pred))

from sklearn.ensemble import RandomForestClassifier
import pprint as pp
import warnings 
warnings.filterwarnings(action='ignore')
import matplotlib as plt
```

```python
load_data = load_breast_cancer() #변수에 집어넣기
```

```python
print(load_data.keys()) # key값 구경
```

```python
df = pd.DataFrame(data = load_data.data, columns = load_data.feature_names)
print(df.head(3))
```

```python
print(df.info())
```

* 이미 데이터가 깨끗함, 전처리 완료된 상태의 데이터이다. 이걸로 모델 학습 평가 해주면 되는 것이다 !

```python
print(load_data.target, load_data.target_names) #['malignant' 'benign']
```

```python
rf_model = RandomForestClassifier(random_state=11) 
#n_estimators = 100 bootstrap을 100번, 트리를 100개, 사용하는 알고리즘은 하나 !
X_train, X_test, y_train, y_test = train_test_split(df, load_data.target, test_size = 0.3, 
                                                    random_state= 11, shuffle=True)
rf_model.fit(X_train,y_train)
pred = rf_model.predict(X_test)
score = accuracy_score(y_test, pred)
print(score)
```

```python
#Python : xgboost pip install
#sklearn : XGBclassfier (sklearn인스톨하면 사용가능한겨)
from xgboost import XGBRFClassifier
# import xgboost as xg
X_train, X_test, y_train, y_test = train_test_split(df, load_data.target, test_size = 0.3, 
                                                    random_state= 11, shuffle=True)
eval_data=[( X_train, y_train)]
# 인코딩
# c.matrix 값을 이용해 다음 분류기 가중치를 준다
# 경사하강
# CV
#파라미터 손대보기
xgb_model = XGBRFClassifier(n_estimators=120, learning_rate=0.1,
                            max_depth=5,random_state = 11)

xgb_model.fit(X_train,y_train, early_stopping_rounds = 30, eval_metric="logloss",
             eval_set=eval_data) #eval_set => 검증 학습 평가 검증 4 : 3 : 3

pred = xgb_model.predict(X_test)
proba = xgb_model.predict_proba(X_test)[:1] #양의 값만 꺼내보겠다는 의미
custom_cls_eval_(y_test, pred,proba)



    #
    #
    #roc_auc_score()
```

```python
#Voting

from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
# ---- 개별 분류 모델 ----
lr = LogisticRegression()
knn = KNeighborsClassifier()
rf = RandomForestClassifier()
xgb_model = XGBRFClassifier(n_estimators=120, learning_rate=0.1,
                            max_depth=5,random_state = 11)

#---- 데이터 셋 ---------------------

X_train, X_test, y_train, y_test = train_test_split(df, load_data.target, test_size = 0.3, 
                                                    random_state= 11, shuffle=True)

#------------앙상블------------------------

vot_model = VotingClassifier(estimators=[("LOR",lr), ("KNN" ,knn), ("RF",rf)], voting = 'soft')


vot_model.fit(X_train,y_train) 

pred = vot_model.predict(X_test)
proba = vot_model.predict_proba(X_test)[:1] #양의 값만 꺼내보겠다는 의미



for mname, model in [("LOR",lr), ("KNN" ,knn), ("RF",rf), ("XGB", xgb_model)] :
    model.fit(X_train,y_train) 
    pred = model.predict(X_test)
#     proba = vot_model.predict_proba(X_test)[:1] #양의 값만 꺼내보겠다는 의미
    print(mname, accuracy_score(y_test, pred))
    

print(custom_cls_eval_(y_test, pred))
    
    

    
    
```
