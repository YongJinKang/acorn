* accuracy_score(y_true, y_pred)

* precision_score(y_true, y_pred)

* recall_score(y_true, y_pred)

* fbeta_score(y_true, y_pred)

* f1_score(y_true, y_pred) precision과 recall 둘다 보는 것으로 볼 수 있기 때문에 accuracy와 가장 많이 쓰는 것 !

* 정밀도 : 정밑으로길게떨어지는도 = TP / FP + TP(암환자라고 예측한 것 중에 맞은 것들)스팸
                     
* 재애애애애애현률 : TP / (FN + TP)   (실제 암환자 중에 제대로 암환자 맞춘 것)   

* 민감도 : (1-재현율)
                  
* accuracy같이 많이 맞출수록, 맞춘 것만 따진다면 ? TN + TP ==> accuracy
 
* TN+TP/TN+FN+FP+TP
 
 
* 모델에 따라, FN이 중요할지 FP가 중요할지, 정밀도가 중요할지, 재현율이 중요할지 다 달라지는구나..!
 
 * 어떤경우는 둘다 중요하다 ! ex)신용카드 사기 So, 정밀도와 재현율 둘다 봐야 한다 ! (f1 score)

                  
                  
 * 정밀도가 올라가면 재현율이 떨어지고, 정밀도가 내려가면 재현율은 떨어진다. trayed-off관계 상극관계

### 정밀도-재현율 곡선 둘다 높아야 좋긴함


```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score
```

```python
print("정확도 ",accuracy_score(y_test3, model_pred))
print("F1:",f1_score(y_test3, model_pred))
print("정밀도: ",precision_score(y_test3, model_pred))
print("재현률(민감도):",recall_score(y_test3, model_pred))
print("ROC:",confusion_matrix(y_test3, model_pred))
print("ROC_AUC:",roc_auc_score(y_test3, model_pred))
```
