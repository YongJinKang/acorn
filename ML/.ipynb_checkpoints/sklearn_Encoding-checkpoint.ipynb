# One Hot Encoding

## Encoding : obj -> int

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder

items = ["R","java","python","python"]
#숫자, 대문자, 소문자 순으로 order by를 내부적으로 진행시킴. 그 순서대로 숫자 배정

encoder = LabelEncoder()
encoder.fit(items)

encoder.transform(items)

label = encoder.transform(items)

print(label, type(label)) #[0 1 2 2] < class 'numpy.ndarray'>
print(encoder.classes_) #원본형태로 돌아가는 것
print(encoder.inverse_transform(label))
```

<class 'scipy.sparse.csr.csr_matrix'>가 무엇인가?
csr_matrix는 희소행렬이다. 저 안에 다 0으로 가득차있는 형태를 희소행렬이라고 한다. onehotencoding은 희소행렬
n차이상의 배열에서 0000값이 주도적으로 많이 보이는 것을 희소행렬이라고 한다.

  (0, 0)	1.0
  
  (1, 1)	1.0
  
  (2, 2)	1.0
  
  (3, 2)	1.0  
  
  이건 위치를 말해주는 것이다 ! (0,0)에 1이있다. (1,1)에 1이 있다 이런식 !


* 행렬을 배열로 바꾸는 방법 toarray(), 리스트는 ? tolist()
: ohlabel.toarray() 

```python
from scipy import sparse
arr = np.array([[0,3,1],[0,6,0],[2,0,0]])
coo = sparse.coo_matrix(arr)
csr = sparse.csr_matrix(arr)
print(coo)
print(csr)
print(coo.toarray())
print(csr.toarray())
```

 * coo와 csr의 차이
    
    coo : 행렬의 위치 좌표 안내, 구동방식이 좀 비효율적 행렬 위치를 다 보여줌
    
    csr : 행의 위치를 중복한 행의 인덱스만 따로 떼서 행번호를 재구성하는 것이 csr

```python
# ex) 0이 아닌 값들의 행 위치
arr = np.array([[0,3,1],[0,6,0],[2,0,0]])
print(arr)
# arr = np.array([[3,1,6,2]])
# row_idx = np.array([0, 0, 1, 2])
# print(row_idx)
# col_idx = np.array([1, 2, 1, 0])
# print(col_idx)
sparse.coo_matrix( arr )
print(type(coo), coo.toarray())
```

## get_dummies()

* 데이터프레임 채로 바꾸거나 할 때 쓰일 수 있음

```python
import pandas as pd
items = ["0python", "R", "java", "0python"]
df = pd.DataFrame(data ={"col" : items })
```

```python
df.head()
```

```python
#본인 프레임에서 인코딩한걸 추가추가하는 모양새가 더미
df = pd.get_dummies((df))
print(df.head())


##데이터 프레임화가 되었다는 전제하에 겟더미를 사용할 수 있는 것이다.
```

### 위의 OneHotEncoding방식으로 위의 데이터프레임 원핫인코딩해보기


```python
encoder = LabelEncoder()
encoder.fit(df.col)

encoder.transform(df.col)

label = encoder.transform(df.col)

col = label.reshape(-1,1)
print(col, type(col))
onehotencoder = OneHotEncoder()
onehotencoder.fit(col)
# onehotencoder.fit(df)

ohlabel = onehotencoder.transform(col)
# ohlabel = onehotencoder.transform(df)

print(ohlabel, type(ohlabel))

print(ohlabel.toarray())
```

```python
# ===================================astype('category')
import pandas as pd
items = ["0python","R","java","0python"]
df = pd.DataFrame(data= {"col" : items})
print(df)
df["col_cate"] = df["col"].astype('category')
print("astype('category'):" , df.head())
```
