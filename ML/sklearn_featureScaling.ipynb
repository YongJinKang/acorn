## 피처 스케일링

* StandardScaler : 평균 0, 분산1 (정규분포를 따름), 아웃라이어에 의해 영향을 많이 받음

* MinMaxSCaler : 0~1 사이에 다 집어넣어라, 아웃라이어에 의해 영향을 많이 받음

* RobustSCaler : 중위수(4분위), 원본데이터 변형 가하지 않음. 중간값을 이용해서 하는데, 아웃라이어 잡기가 편함

* 


```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import RobustScaler
from sklearn.datasets import load_iris
import pandas as pd
```

```python
iris_data = load_iris()
data = iris_data.data
df =pd.DataFrame(data = data, columns= iris_data.feature_names)
print(df.head())
```

```python
df.describe() 
```

```python
print(df.describe() ,"\n")
print(df.mean(),"\n",df.var(),"\n")
```

# --------------------------Standard Scaler--------------------------

```python
std_scaler = StandardScaler()
std_scaler.fit(df)
std_scaler.transform(df)
ndarray_scaler = std_scaler.transform(df)       #[[][][]...]
sdf = pd.DataFrame(data = ndarray_scaler, columns=iris_data.feature_names)
print(sdf.mean(), sdf.var(), "\n")
```

# --------------------------MinMax Scaler--------------------------

```python
df = pd.DataFrame(data = data, columns=iris_data.feature_names)
mscaler = MinMaxScaler()
mscaler.fit(df)
ndarray_scaler = mscaler.transform(df)
#데이터프레임으로 변환하여서 확인
sdf = pd.DataFrame(data=ndarray_scaler, columns=iris_data.feature_names)
print(sdf.mean(), sdf.var(), "\n")
```

# --------------------------RobustScaler--------------------------

```python
df = pd.DataFrame(data = data, columns=iris_data.feature_names)
rscaler = RobustScaler()
rscaler.fit(df)

ndarray_scaler = rscaler.transform(df)
sdf = pd.DataFrame(data = ndarray_scaler, columns = iris_data.feature_names)
print(sdf.mean(),"\n" ,sdf.var() ,"\n")
```

* scale조정 전에 아웃라이어 제거해야 한다 !

# ==================fit() transform() fit_transform()

```python
#fit() : 변환을 준비, 학습시키는 것, 스케일을 (데이터 변환을 학습시킨다.)
#transform() : 실제 스케일 조정
import numpy as np
X_train = np.arange(0,11).reshape(-1,1) #0 ~ 10
print(X_train)
print("원본 : " , X_train.reshape(-1))
y_train = np.arange(0,6).reshape(-1,1) # 0 ~ 5
print(y_train)
print("원본 : " , y_train.reshape(-1))
```

```python
scaler = MinMaxScaler()
scaler.fit(X_train) #1/10
X_train_scaler = scaler.transform(X_train)
print(X_train_scaler.reshape(-1))
```

```python
scaler.fit(y_train) #1/5
y_train_scaler = scaler.transform(y_train)
print(y_train_scaler.reshape(-1))
```

스케일이 조정되고 났으면 조정된거로만 할 것이다. 우리가 그럼 0.4라고 하면 값의 수치가 원래 뭘까 ?4 ? 2?
핏핏두번하니까  각자 스케일을 시켜버리니까 달라버려진다.. 이러면 비교가 안돼 ! 같이 묶어서 해야 하는 거야 !
하나는 1/5 하나는 1/10으로 표준화를 시켜서 이상하게 값이 나와버리는 것이다 !


* 해결하는 방법: 핏 한번만 하면 된다 !

```python

X_train_scaler = scaler.fit_transform(X_train)
print("0~10:",X_train_scaler.reshape(-1))

y_train_scaler = scaler.fit_transform(y_train) #1/5
print("0~5:",y_train_scaler.reshape(-1))
```

```python
scaler.fit(X_train)#-------------------------1/10
X_train_scaler = scaler.transform(X_train)
print("1/10: ",X_train_scaler.reshape(-1))
# scaler.fit(y_train) #1/5
y_train_scaler = scaler.transform(y_train)
print("1/10: ",y_train_scaler.reshape(-1))
```

* 이거 정말 많이 하는 실수이다 ! fit은 한번만 하여서 (따로따로 하지말고 !) 동일한 표준화를 적용하여

*  두값을 비교할 필요가 있다 !
